```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler

# === 1. Carica il tuo dataset pulito ===
df = pd.read_excel(
    "/Users/Altair93/Documents/Dottorato/PATHS/Python_analisys/Intermediate_data/statistic/Database_PlanB.xlsx"
)

# === 2. Seleziona le colonne EEG di interesse (es. solo globali) ===
features = [
    "delta_avg",
    "theta_avg",
    "exponent_avg",
    "higuci_avg",
    "lziv_avg",
]
X = df[features].copy()

# === 3. Standardizza le variabili (importantissimo per VIF) ===
X_scaled = pd.DataFrame(StandardScaler().fit_transform(X), columns=X.columns)

# === 4. Calcola i Variance Inflation Factors ===
vif_data = pd.DataFrame()
vif_data["Feature"] = X_scaled.columns
vif_data["VIF"] = [
    variance_inflation_factor(X_scaled.values, i)
    for i in range(X_scaled.shape[1])
]

print("=== Variance Inflation Factors (VIF) ===")
print(vif_data.round(3))

# === 5. Soglia interpretativa ===
# VIF < 2  → nessuna collinearità
# VIF 2–5  → moderata (accettabile)
# VIF > 5  → alta (da evitare)
```
